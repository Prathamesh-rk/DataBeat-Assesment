```python
import pandas as pd
import json
import ast
import numpy as np
```

```python
activity_data = pd.read_csv('activity_data.csv')
activity_data
```

```python
activity_data["_id"] = [int(ast.literal_eval((i.replace("'", '"')))["S"].split("_")[0]) for i in activity_data["id"]]
activity_data
```

```python
temp = activity_data[["_id", "rating", "hide", "watchlist", "watched"]].copy()

for i in temp:
  if i == "_id":
    continue
  temp[i] = [ast.literal_eval((j.replace("'", '"')))["M"]["value"].get("S", np.nan) for j in temp[i]]
temp
```

```python
data = pd.read_csv('data.csv')
data
```

```python
temp2 = pd.merge(data, temp, how='left', on = "_id")
```

```python
temp2
```

```python
lst = []
for i in temp2.itertuples():
  lst.append(f"{i[4][:-1]}, 'rating': {i[5]}, 'hide': {i[6]}, 'watchlist': {i[7]}, 'watched': {i[8]}"+"}")
```

```python
res = pd.DataFrame(lst, columns = ["_source"])
res
```

```python
res.to_csv("Python_Assessment_Output.csv", index = False)
```
